# Binary Search

Given a list of numbers or elements, search a target element
- requires list to be sorted 
- we will use L, R, and Mid pointers for setting up boundaries for searching
- L - left/low, R - right/high
- mid = (L+R)//2 or mid = low + (high-low)//2  # to avoid overflowing

You are given an array of distinct integers nums, sorted in ascending order, and an integer target.
Implement a function to search for target within nums. If it exists, then return its index, otherwise, return -1.

Your solution must run in

O(logn) time.

Example 1:

- Input: nums = [-1,0,2,4,6,8], target = 4

- Output: 3
Example 2:

- Input: nums = [-1,0,2,4,6,8], target = 3

- Output: -1
Constraints:

1 <= nums.length <= 10000.
-10000 < nums[i], target < 10000


```python
class BinarySearch:
    def __init__(self, nums, target):
        self.nums = nums
        self.target = target
    
    def recursive_binary_search(self, l, r):
        '''
        Time complexity: O(n log n)
        Space Complexity: O(1)
        '''
        
        mid = l + (r-l)//2
        if self.nums[mid] > self.target:
            return self.recursive_binary_search(l, mid-1)
        elif self.nums[mid] < self.target:
            return self.recursive_binary_search(mid+1, r)
        else:
            print(f"printing mid {mid}")
            return mid
    
    def iterative_binary_search(self, l, r):
        l, r = 0, len(self.nums)-1
        if l > r:
            return -1

        while l <= r:
            m = l + ((r-1)//2)
            if nums[m] > target:
                r = m-1
            elif nums[m] < target:
                l = m + 1
            else:
                return m
        
    def search(self):
        l, r = 0, len(self.nums)-1
        if l > r:
            return -1
        return self.recursive_binary_search(l, r)
```

```python
x = BinarySearch(nums = [-1,0,2,4,6,8], target = 4)
print(x.search())
```
