### Factorials
1. Lets start with an example: 5! = 5 * 4 * 3 * 2 * 1
2. or we can also represent it as 5! = 5 * 4!
3. or 5! = 5 * 4 * 3!
4. General Formula n! = n * (n-1)!

#### One Branch Recursion
1. Recursion function is when a function calls itself. usually with a different input.
2. Recursion function breaks down problem into smaller sub-problem and solves them in reverse order.
3. Every iterative function can be converted to recursive one and vice versa.


### Recursion has two parts
1. The base case (5! can be broken to 5 *4!, similarly 4! = 4*3! erc)
2. The recursive case (0! or 1! which cannot be further broken down)

```python
def factorial(n):
    if n <= 1:
        # base case
        return 1
    # recursive case
    return n * factorial(n-1)
```

```python
print(factorial(5))
```
