#### Reverse a linked list 

```python
class LinkedList:
    def recursive_reverse(self, head):
        if not head:
            return None
        
        new_head = head
        if head.next:
            newHead = self.reverseList(head.next)
            head.next.next = head
        head.next = None
        
        return newHead
```

```python
class Solution:
    def reverseList(self, head):
        prev, curr = None, head

        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp
        return prev
```
