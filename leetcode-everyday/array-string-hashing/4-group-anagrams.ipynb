# Group Anagrams
Given list of strings, return a list grouping anagrams
```python
# Example 1
Input: strs = ["act","pots","tops","cat","stop","hat"]
Output: [["hat"],["act", "cat"],["stop", "pots", "tops"]]
```
```python
# Example 2
Input: strs = ["x"]
Output: [["x"]]
```
```python
# Example 3
Input: strs = [""]
Output: [[""]]
```

````python
class GroupAnagrams:
    def __init__(self, words):
        self.words = words
    
    def indexHashing(self):
        from collections import defaultdict
        word_dict = defaultdict(list)
        for word in self.words:
            count = [0] * 26
            for ch in word:
                count[ord(ch) - ord('a')] += 1
            word_dict[tuple(count)].append(word)
        
        return word_dict.values()
x = GroupAnagrams(["act","pots","tops","cat","stop","hat"])
print(x.indexHashing())
            
````
