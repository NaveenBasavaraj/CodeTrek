# Anagram Problems

1. Check if two given strings are valid anagrams
    ```python
    # Example 1
    Input: s = "racecar", t = "carrace"
    Output: true
    ```
    ```python
    # Example 2
    Input: s = "jar", t = "jam"
    Output: false
    ```


````python
class IsAnagram:
    def __init__(self, s1, s2):
        self.s1 = s1
        self.s2 = s2
    
    def hashing(self):
        if len(self.s1) != len(self.s2):
            return False
        count1, count2 = {}, {}
        
        # return sorted(self.s1) == sorted(self.s2)
        
        for i in range(len(self.s1)):
            count1[self.s1[i]] = 1+count1.get(self.s1,0)
            count2[self.s2[i]] = 1+count2.get(self.s2,0)
        return count1 == count2
    
    def indexHashing(self):
        if len(self.s1) != len(self.s2):
            return False
        count = [0] * 26
        for i in range(len(self.s2)):
            count[ord(self.s1[i]) - ord('a')] += 1
            count[ord(self.s2[i]) - ord('a')] -= 1
        
        for val in count:
            if val != 0:
                return False
        return True

x = IsAnagram("racecar", "carrace")
print(x.hashing())
print(x.indexHashing())
````
