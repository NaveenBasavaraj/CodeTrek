# Binary Trees
### A Connected, Undirected Graph With No Cycles

- Binary trees are data structures
- which involves **nodes** and **pointers** similar to linked list
- instead of **next** or **prev** we have **left_child** and **right_child**
- uses **TreeNode** datatype similar to **ListNode** in a linked list
- unlike linked list, trees do not allow cycles

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# class ListNode:
#     def __init__(self, value):
#         self.value = value 
#         self.next = None
#         self.prev = None
```

## Tree Terminologies

1. **Root Node:**
    - highest node in the tree with no parents
    - any child node can be reached from the tree

2. **Leaf Node:**
    - a node without any children
    - nodes at the last level of the tree

3. **Ancestor**: A node is an ancestor to all the child nodes below it
4. **Descendent**: the descendent of a node is either the child node or child of some other descendant of the node

5. **Height**:
    - Number of nodes from root to the lowest leaf node
    - n-1 where n is the number of nodes from the root to the lowest lead node
    - we can also use edges between nodes instead of nodes itself
    - number of edges -> n-1
    - if root itself is counted its +1 else +0
    - height of single tree node is 1 if root is considered else 0

6. **Depth**:
    - Depth of a binary tree node is measured from itself all the way up to the root
    - Measure depth at a given node by looking at how many nodes are above it, including the node itself




## Types Of Binary Tree

1. **Full Binary Tree**: Each node has either 2 or 0 children.
2. **Perfect Binary Tree**: Each node has exactly two children
3. **Complete Binary Tree**: 
    - Every level is completely filled
    - tree nodes must be left leaning
    - last leaf node need not contain right child
4. **Pathological or Degenerate**: Each node contains only left or right child
5. **Skewed Binary Tree**: dominated bye either right nodes or left nodes
    - Left skewed: only contains left children
    - Right Skewed: only contains right children
