```python
# Magic methods that turn an object into a sequence
class MyCustomSequence:
    '''
    A sequence:
        1. supports indexing
        2. stores objects
        3. supports len() function
    '''

    def __init__(self, data):
        self._data = data
    
    def __len__(self):  # Tells Python how long the sequence is
        return len(self._data)
    
    def __getitem__(self, index):  # Allows indexing and slicing
        return self._data[index]
    
    def __iter__(self):  # Makes the object iterable
        return iter(self._data)

# Now you can use your custom sequence like a list!
magical_seq = MyCustomSequence([1, 2, 3, 4])
print(len(magical_seq))  # 4
print(magical_seq[2])    # 3
for item in magical_seq: # Works in a loop!
    print(item)
```
