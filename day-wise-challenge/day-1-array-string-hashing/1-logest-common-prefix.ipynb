# LONGEST COMMON PREFIX

- Write a function to find the longest common prefix string amongst an array of strings.
- If there is no common prefix, return an empty string "".

### Example 1
- Input: strs = ["flower","flow","flight"]
- Output: "fl"

### Example 2
- Input: strs = ["dog","racecar","car"]
- Output: ""
- Explanation: There is no common prefix among the input strings.

```python
from typing import List
class LongestCommonPrefix:
    def __init__(self, strings:List[str]) -> str:
        self.strings = strings
    
    def horizontalScan(self):
        '''
        time: o(m*n)
        space: o(n)
        '''
        prefix = self.strings[0] # first letter from first string
        for i in range(1, len(self.strings)):
            j = 0
            while j < min(len(prefix), len(self.strings[i])):
                if prefix[j] != self.strings[i][j]:
                    break
                j += 1
            prefix = prefix[:j]
        return prefix

solution = LongestCommonPrefix(["flow", "flower", "flight"])
print(f" longest prefix is {solution.horizontalScan()}")
```
